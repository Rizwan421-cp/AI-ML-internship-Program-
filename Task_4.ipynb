```python
from openai import OpenAI

# Replace with your actual API key
client = OpenAI(api_key="Your_API_KEY")

SYSTEM_PROMPT = """
You are a friendly and knowledgeable health assistant.

Your task is to answer general health questions clearly and simply.
Do not diagnose diseases or prescribe medication.
If the question seems serious or urgent, advise the user to seek medical help.

Keep answers short and easy to understand.
Always end with: "Consult a doctor for personal medical advice."
"""

conversation_history = [{"role": "system", "content": SYSTEM_PROMPT}]
MAX_HISTORY = 10

def is_sensitive(query):
    keywords = ["suicide", "overdose", "emergency", "severe pain", "chest pain"]
    return any(word in query.lower() for word in keywords)

def chat(user_message):
    if is_sensitive(user_message):
        return "This may be serious. Please contact a doctor or emergency service immediately."

    conversation_history.append({"role": "user", "content": user_message})

    while len(conversation_history) > MAX_HISTORY + 1:
        conversation_history.pop(1)
        conversation_history.pop(1)

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=conversation_history,
        temperature=0.5
    )

    reply = response.choices[0].message.content
    conversation_history.append({"role": "assistant", "content": reply})

    return reply


# Run chatbot
print("Health Assistant ready. Type 'quit' to exit.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() == "quit":
        print("Goodbye!")
        break

    response = chat(user_input)
    print("Assistant:", response)
```
